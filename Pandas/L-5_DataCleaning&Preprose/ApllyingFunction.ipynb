```python
 import pandas as pd
df= pd.read_excel("data.xlsx.xlsx")
print(df)
```

```python
# ****Most important topic in pandas 
# apply()
def isGood(x):
    return "Good" if x>=15000 else "below"

df["Member"]= df["Salary"].apply(isGood)
# df["Member"]= df["Salary"].apply(lambda x: "Good" if x>=15000 else "below")

print(df)
```

```python
# 2. Map()
gender_Map={  "M ":"Male" , "    F": "Female"}
# map() change all value 
df["Gender"] = df["Gender"].replace(gender_Map)
print(df)
```

```python
# .replace()

df["Name"] =df["Name"].replace({ "Sunil ":"Hero" , "Noori":"N"})
print(df)
```
