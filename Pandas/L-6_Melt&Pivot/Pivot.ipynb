```python
# Most important nethods 
# 1.   melt()   -- Wide to long 
import pandas as pd
data = {
    "Name" : ["Alice" ,"Alice" ,"Bob","Charlie"],
    "Math": [85,78,92,34],
    "Science" :[90,32,34,34],
    "English" :[88,95,94,34]
}

df =pd.DataFrame(data)
print(df)


```

```python
df1=df.melt(id_vars=["Name"] ,value_vars=["Math","Science","English"] ,var_name="Subject", value_name="Score").copy()
print(df1)
```
