```python
import pandas as pd
import pandas as pd

# List of dictionaries format
dat1 = {
    "Department": ["HR", "HR", "IT", "Member", "Member", "Member", "Leader", "Leader", "Sales"],
    "salary":    [50000, 40000, 70000, 40000, 20000, 10000, 90000, 50000, 55000],
    "age":       [40, 35, 30, 21, 20, 28, 40, 20, 32]
}


# Create DataFrame directly
df = pd.DataFrame(dat1)

# Display DataFrame
print(df)
```

```python
df.groupby("Department")["salary"].mean()
# df.groupby(["Department","age"])["salary"].mean()
```

```python
df.groupby("Department")["salary"].count()  # com
# df.groupby("Department")["salary"].max()  # com
# df.groupby("Department")["salary"].min()  # give min of salary 
# df.groupby("Department")["age"].min()  # give min of salary 
```

```python
# we can custom all above function in on line 
df.groupby("Department")["salary"].agg(["mean","max","min","count"])


# we can also give special for all aggregation column 
df.groupby("Department")["salary"].agg(
    avg_sal="mean",high_score="min"
    

)
```

```python
# transform funtion in pandas
df["Higher"]=df.groupby("Department")["salary"].transform("max")
df
```

```python
# filter elemnt based on mean,max and etc.
df.groupby("Department").filter(lambda x:x["salary"].mean()>50000)
```
