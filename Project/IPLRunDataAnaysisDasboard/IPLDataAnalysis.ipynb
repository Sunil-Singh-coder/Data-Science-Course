```python
import pandas as pd

df= pd.read_csv("IPL_Ball_by_Ball_2022.csv")
# print(df.head())
df1=df.query("ID == 1312200")
print(df1)
```

```python
# Bar Chart → Top Scorers in matches.
import matplotlib.pyplot as plt
batter=df1["batter"].drop_duplicates().reset_index(drop=True).copy()
score=df1.groupby("batter")["total_run"].sum().tolist()
batter=batter.tolist()
newbatter= [i[0:4] for i in batter] 
plt.bar(newbatter,score)
plt.tight_layout()
plt.grid(True)
# plt.figure(figsize=(15,5))
plt.title("Top Scorers in matches.")
plt.xlabel("Batter name")
plt.ylabel("Run")


```

```python
# Line Chart → Runs scored over time.
inning1= df1.query("innings==1")
ballerseries=inning1.groupby("bowler")["total_run"].sum()
bolwer_name=ballerseries.index.tolist()
bolwer_run=ballerseries.values.tolist()
bolwer_names= [i[0:9] for i in bolwer_name] 
plt.bar(bolwer_names,bolwer_run)
plt.title("Top bolwer run  in matches.")
plt.xlabel("bolwer name")
plt.ylabel("Run")

# inning1.groupby("innings")["total_run"].sum()

```

```python
# Histogram → Distribution of Runs.
# Histogram shows how many times runs fall into a certain range (bin).
inning1= df1.query("innings==1").copy()
# runseries=inning1.groupby("overs")["total_run"].sum()
# run_bu_over= runseries.values.tolist()
# print(run_bu_over)
# plt.hist(run_bu_over,bins=[0,5,10,15,20],edgecolor="black")
```

```python
bins = [0, 5, 10, 15,20]
labels = ['0-5 Overs', '5-10 Overs', '10-15 Overs','15-20 Overs']
inning1['over_group'] = pd.cut(inning1['overs'], bins=bins, labels=labels, right=False)
run_by_group= inning1.groupby("over_group", observed=True)["total_run"].sum()

group=run_by_group.index.tolist()
run=run_by_group.values.tolist()
plt.bar(group,run)
```
