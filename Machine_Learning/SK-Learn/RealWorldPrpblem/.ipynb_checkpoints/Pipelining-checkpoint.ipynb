```python
from sklearn.model_selection import StratifiedShuffleSplit
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("housing.csv")
```

```python
split = StratifiedShuffleSplit(n_splits=1,test_size=0.1,random_state=42)
for train_index,test_index in split.split(df,df["ocean_proximity"]):
    train_set=df.iloc[train_index]
    test_set=df.iloc[test_index]
# train_set
# test_set
```

```python
housing=   train_set.drop("median_house_value",axis=1)
housing_lables=  test_set["median_house_value"]
```

```python
# We use option 3 by using Simple Imputer 
from sklearn.impute import SimpleImputer 
imputer = SimpleImputer(strategy="median")
housing= housing.select_dtypes(include=[np.number])
imputer.fit(housing)
housing["ocean_proximity"]= train_set["ocean_proximity"]
```

```python
# now we are going to encode by ONE Hot Encoder 
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
housing["ocean_proximity"]= train_set["ocean_proximity"]
housing_cat= housing[["ocean_proximity"]]           # work on ocean proximity 
housing_cat_hot_encoded= onehotencoder.fit_transform(housing_cat)
housing_cat_hot_encoded.toarray()
housing_cat_df= pd.DataFrame(housing_cat_hot_encoded.toarray(),columns=['<1H OCEAN', 'INLAND', 'ISLAND', 'NEAR BAY', 'NEAR OCEAN'],index= housing.index)

housing=housing.drop("ocean_proximity",axis=1)
housing = pd.concat([housing, housing_cat_df],axis=1)
```

```python
#  now we are going to sacle the large value to small value for fast process data
# we have 2 way to solve this problem 
 # 1---> MinMaxScaler        2------> StandardScaler

# 1 MinMaxScaler   
     # scaled_value= (x-min)/(max-min)

from sklearn.preprocessing import MinMaxScaler 
scaler= MinMaxScaler(feature_range=(-1,1))
housing_scale_minmax= scaler.fit_transform(housing)
housing_scale_minmax_df= pd.DataFrame(housing_scale_minmax,columns=housing.columns,index= housing.index)
housing_scale_minmax_df
```

```python
# way 2 in case of wrong data we 

# 2---> StandardSacler()
from sklearn.preprocessing import StandardScaler
Sscaler= StandardScaler()
housing_standard= Sscaler.fit_transform(housing)
housing_standard_df= pd.DataFrame(housing_standard,columns=housing.columns,index= housing.index)
housing_standard_df
```
