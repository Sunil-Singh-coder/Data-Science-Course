```python
# 1.Write code to split the array [10,20,30,40,50,60] into 3 equal parts.
import numpy as np
arr= np.array([10,20,30,40,50,60])
arr1=np.split(arr,3)
print(np.split(arr,3))
print(arr1)
```

```python
# Generate a 2x4 matrix with random integers between 1 and 20.
arr= np.random.randint(1,20,8).reshape(2,4)
print(arr)
```

```python
# 3.Write NumPy code to vertically stack:
# [1,2,3] 
# [4,5,6] 
arr1= np.array([1,2,3] )
arr2= np.array([4,5,6])
print(np.vstack([arr1,arr2]))
print(np.hstack([arr1,arr2]))
```
