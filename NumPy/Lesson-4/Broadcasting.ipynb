```python
# Now, we’ll explore how to make your code faster
# with vectorization and broadcasting in NumPy.
# These techniques are key to boosting performance 
# in numerical operations by avoiding slow loops and memory inefficiency.
import numpy as np

# 1 vectorization 
arr = np.array([2,32,2,1])
arr=arr**2
print(arr)


```

```python
# Scaling Arrays Without Extra Memory
arr= np.array([2,1,2,1])
arr= arr+10   # broadcasting 10 is added to all element 
print(arr)
```

```python
# . Broadcasting with Arrays of Different Shapes

# arr = np.array([2,32,2,1])
# arr1 = np.array([2,3,3,2])
# arr= arr+arr1
# print(arr)

arr2D = np.array( [[2,32,2,1] , [2,3,3,2]])
arr1 = np.array([2,3,3,2])
arr= arr2D+arr1
print(arr)
```

```python
# Hands-on: Applying Broadcasting to Real-World Scenarios

# ----> Normalizing Data Using Broadcasting
arr= np.array([
       [5,5,2],
       [2,4,2],
       [8,3,8],
       [1,7,2]  
])
mean = arr.mean(axis=0)
std= arr.std(axis=0)
normalized = (arr-mean)/std
print(normalized)
# •	Broadcasting enables operations between arrays
# of different shapes by automatically stretching the smaller array to 
# match the shape of the larger array, without creating additional copies.

```
