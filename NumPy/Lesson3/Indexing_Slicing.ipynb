```python
import numpy as np

arr= np.arange(1,10)
flat = arr.flatten()
print(flat)
print(flat)




```

```python
# slicing performed as a list 
print(flat[3])
print(flat[3:5])
print(flat[3:15])
print(flat[3:])
print(flat[:])
```

```python
#  slicing return a view not a copy ! change affect the orginalarray
b= flat[3:6].copy()   # [1, 2, 3, 4, 5, 6, 7, 8, 9]
b[0]=234        #[4,5]
print(flat)   # [  1   2   3 234   5   6   7   8   9] changed
```

```python
#  to make a copy of array we use np.copy
b= flat[3:6].copy()  # [1, 2, 3, 4, 5, 6, 7, 8, 9]
b[1]=400     #[4,5]
print(flat)   #[1 2 3 4 5 6 7 8 9]  not changed
print(b)       #[  4 400   6]

```

```python
# Fancing indexing 
arr= np.array([2,32,24,65,88,87,56,54])
```

```python
index=[2,3,4,5]  
print(arr[index])   #[24 65 88 87]  fancing indexing
```

```python
# boolean masking (filter data)
mask =arr>30
arr[mask]  #array([32, 65, 88, 87, 56, 54])
arr[arr>50]   # array([65, 88, 87, 56, 54])
```
