```python
from bs4 import BeautifulSoup
```

```python
with open("html/page1.html1",encoding="utf-8") as f:
    content=f.read()
```

```python
soup= BeautifulSoup(content,"html.parser")
```

```python
h3s=soup.find_all("div")
h3s=soup.find_all("p")
h3s=soup.find_all("h1")
h3s=soup.find_all("h2")
print(h3s)
```

```python
list=[]
quotes=soup.select("div.quote")
for quote in quotes:
    author=quote.select_one("small.author").text
    tags=quote.select_one("div.tags").find("a").text
    
    list.append([author,tags])
    # print(tags)
print(list)
    
    # print(list(tags))
    # print(author)
    

```
