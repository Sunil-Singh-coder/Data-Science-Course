```python
import matplotlib.pyplot as plt
```

```python
years= [2000,2001,2002,2003,2004,2005,2006]
runs= [100,230,300,200,500,300,100]
plt.bar(years,runs)
plt.xlabel("years")
plt.ylabel("runs")
```

```python
# Setting width and side by side by side bar chart 
import numpy as np
Sachin= [500,600,200,700,400,200,100]
virat= [0,233,333,200,300,500,200]
kohli = [0,0,200,300,400,200,100]
x=np.arange(len(years))
width =0.25
plt.bar(x - width,Sachin,width=width,label="Sachin")
plt.bar(x,virat,width=width,label="virat")
plt.bar(x + width,kohli,width=width,label="kohli")
plt.xticks(x,years)

plt.legend()

plt.show()
```

```python
# Horizonatal chart with barh()

players= ["Sunil:" ,"Rohan", "Anshu"]
years= [100,215,201]
plt.barh(players,years,color="red")
plt.legend(["kohli"])
for i in range(len(players)):
    plt.text(i,years[i]+50,str(years[i]),ha="center")
```
