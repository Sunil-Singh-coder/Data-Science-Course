```python
import matplotlib.pyplot as plt 
```

```python
x=[1,2,3,4,5,6,7,8,9]
y1= [i*2 for i in x]
y2= [i*i for i in x]
y3= [0.5*i  for i in x]
y4= [10-i for i in x]
```

```python
fig,axs= plt.subplots(2,2,figsize=(5,5))
ys= [y1,y2,y3,y4]
titles= [" double" , " square","0.5*i" , "10-i"]
for i in range(2):
    for j in range(2):
        idx= i*2+j;
        axs[i,j].plot(x,ys[idx])
        axs[i,j].set_title(titles[idx])
plt.tight_layout()        
```
