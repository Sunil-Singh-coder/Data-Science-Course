```python
import matplotlib.pyplot as plt 
```

```python
# subplot alllow you to show multiple plot in single figure 

# im this example we plot 2 plot 1 is double of number and 2 is square of number 
x=[1,2,3,4,5,6,7,8,9]
x1= [i*2 for i in x]
x2= [i*i for i in x]
y3= [0.5*i  for i in x]
y4= [10-i for i in x]
plt.figure(figsize=(7,5))

plt.subplot(2,2,1)
plt.plot(x,x1)
plt.title("Double of number ")

# plt.figure(figsize=(5,3))
plt.subplot(2,2,2)
plt.plot(x,x2)
plt.title("Square of number ")


plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("Dossr ")

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("Doubmber ")

plt.tight_layout()
plt.show()
```
