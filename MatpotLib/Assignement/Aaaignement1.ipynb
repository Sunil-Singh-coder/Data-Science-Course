```python
# problem : -->> 
# Create a pie Chart showing the market share of mobile os 
# Andorid,ios,others) Then Create the same using horizonatal
# bar chart and observe which is easier to understand

import matplotlib.pyplot as plt 
MobileOs = ["Andorid" ,"iOS" , " other"]
data= [72.2,26.9,0.9]
explode=[0,0,0.1]
plt.pie(data,labels=MobileOs,shadow=True,autopct="%1.1f%%",
       explode=explode
       )






```

```python
plt.figure(figsize=(18,20))
plt.barh(MobileOs, data, color=['green', 'blue', 'gray'])
plt.xlabel('Market Share (%)')
plt.title('Mobile OS Market Share (Horizontal Bar Chart)')
plt.show()
```
