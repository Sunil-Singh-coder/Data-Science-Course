```python
import matplotlib.pyplot as plt
```

```python
# 
run =[2000,3000,5000,1000]
name =["Sunil","Rohan","Anshu","Afreen"]
color=["red","green","yellow","pink"]
# using wedgeprops to set Edges
plt.pie(
    run,labels=name,
    colors=color,
    wedgeprops= { 'edgecolor': 'black','linewidth':3,'linestyle':'--'}
    
)
```

```python
explode=[0.1,0,0,0]
plt.pie(
    run,labels=name,
    colors=color,
   
    explode=explode
    
)
```

```python
# shadow is use show 3D model 
plt.pie(
    run,labels=name,
    colors=color,
    explode=explode,
    shadow=True  
)
```

```python
# To rotate our pie chart 
plt.pie(
    run,labels=name,
    colors=color,
    explode=explode,
    shadow=True,  
    startangle=180
)
```

```python
# do comparison through percentage we use autopct 
plt.pie(
    run,labels=name,
    colors=color,
    explode=explode,
    shadow=True,  
    startangle=180,
    autopct='%1.1f%%'
)
```
