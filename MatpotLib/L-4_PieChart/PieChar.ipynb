```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

```python
run =[2000,3000,5000,1000]
name =["Sunil","Rohan","Anshu","Afreen"]
# plt.pie(run,name=name
plt.pie(run, labels=name)
# plt.pie(run, labels=name, autopct='%1.1f%%')
plt.show()
```

```python
run =[2000,3000,5000,1000]
name =["Sunil","Rohan","Anshu","Afreen"]
# plt.pie(run,name=name
color=["red","green","yellow","pink"]
plt.pie(run, labels=name,colors=color)
# plt.pie(run, labels=name, autopct='%1.1f%%')
plt.show()
```
