```python
# Now in stack plot
days=[1,2,3,4]
wactching=[0,3,5,2]
sleeping=[5,1,5,2]
playing=[6,1,3,2]
study=[3,2,1,2]
colors= ["red","pink","Green","blue"]
activities =["Studing" ,"Playing","Watching","Sleeping"]
plt.stackplot(days,study,playing,wactching,sleeping,
              alpha=0.8
              ,labels=activities,)
plt.legend()
plt.xlabel("Days")
plt.ylabel("Hours")
plt.title("Weekly schedule")
plt.grid(True)
```

```python
import matplotlib.pyplot as plt

# Data
days = [1, 2, 3]
math = [2, 3, 4]
science = [3, 2, 1]
english = [1, 2, 3]

# Stack plot
plt.stackplot(days, math, science, english, labels=['Math', 'Science', 'English'], colors=['blue', 'green', 'orange'])
plt.legend(loc='upper left')
plt.title('Study Time Distribution Over Days')
plt.xlabel('Days') 
plt.grid(True)
plt.ylabel('Hours Studied')
plt.show()
```
