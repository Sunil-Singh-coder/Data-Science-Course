```python
import matplotlib.pyplot as plt

# use karte hai jab hame kisi data ko
# variable se comparison karna hota hai tab 
```

```python
# Basic Scatter plot 
study=[1,2,3,4,5,6,7]
scores=[20,10,15,25,45,35,40]
# plt.scatter(study,score)
plt.grid(True)
plt.xlabel("Hours")
plt.ylabel("Score")
plt.title("Score VS Hours")  
c=['red' if score < 25 else "green" for score in scores ]
s=[5*score for score in scores]
plt.scatter(study,score,c=c,s=s,cmap='viridis')
```

```python
# we can use c and s for color and size respectively in scatter function 
# Using colormap 
study=[1,2,3,4,5,6,7]
scores=[20,10,15,25,45,35,40]
# plt.scatter(study,score)
plt.grid(True)
plt.xlabel("Hours")
plt.ylabel("Score")
plt.title("Score VS Hours")  
c=['red' if score < 25 else "green" for score in scores ]
s=[5*score for score in scores]
plt.scatter(study,score,c=c,s=s)
for i in range(len(score)):
    plt.annotate(f'S{i} ',(study[i]+0.2,scores[i]-0.5))
```
